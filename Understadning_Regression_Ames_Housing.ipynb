#1.2 Simple Linear Regession
<font size = 12 face = leelawadee>
Simple Linear Regession <br>
$ y = w_1x + b$
$\text {b is bias unit or y axis intercept}$
$\text {w is weight or coefficent or parameter}$
$\text {x is the explanatory or independent var or feature}$
$\text {simple means there is only one x one is explanatory or feature}$
$\text {Goal : Learn the weight of the linear eqn that describes the relationship between explanatory and target }$
$\text {Knowing the value of weights w and b helps predict y when have new (unseen) values of x }$

|Variable| Name |Another Name
--------|--------|--------|
|y| Target |Response or dependent|
|x| Feature |Explanatory or Independent|



#2.1 Line of best of Fit <br>
<font size = 14 face = leelawadee>
Line of best of Fit : Simple Example

```python
x = np.array([1,2,3,4,5])
y = np.array([3.1,2.8,8.9,6.1,13.9])
plt.figure(figsize = (15,10))
plt.plot(x,y,'o')
plt.xlabel('x : Explanatory', size = 26, color = 'green' )
plt.ylabel('y : Response', size = 26, color = 'green' )
w, b = np.polyfit(x,y, deg =1 ) # compute slope and intercept
plt.axline(xy1=(0, b), slope = w, color = 'red', lw = 3) # plot slope and intercept
plt.ylim([-5, 20])
plt.xlim([-5, 20])
plt.axvline(0, color ='green')
plt.axhline(0, color ='green')
plt.text(x= 7, y = 15, s = 'Line of best fit or\n Regression Line', \
         color = 'red', size = 20)

# show residuals for a couple of points for clarity
plt.vlines(x[3], y[3], w*x[3] + b, color = 'green', ls = 'dashed')
plt.vlines(x[4], y[4], w*x[4] + b, color = 'green', ls = 'dashed')
plt.text(x = 0.1*x[4], y = 1.05*(w*x[4] + b), s= 'Dashed line shows \n residual or \n error in prediction or \n vertical offset \n $|y_{pred} - y|$ ', size = 14, color = 'green')
plt.title(f'Plot of \n $y = {w:.2f}*x + {b:.2f}$', size = 28)
plt.show();

#  (2.4899999999999993, -0.5099999999999989)
```

#3.2 Correlation
<font size = 18 color = green face = leelawadee>
Correlations can range between -1 and +1 <br>.
Pearson’s correlation coefficient can simply be calculated as the covariance between two
features, x and y (numerator), divided by the product of their standard deviations (denominator):<br>
<font size = 40 color = darkred>
$\text{Correlation }r = \frac{\Sigma_{i=1}^n [(x^i - \mu_x) (y^i - \mu_y)]}{\sigma_x \sigma_y} = \frac{\sigma_{xy}} {\sigma_x \sigma_y}$
$\sigma_x = \sqrt{(x^i - \mu_x)^2}$ <br>
$\sigma_y = \sqrt{(y^i - \mu_y)^2}$ <br>
$\sigma_{xy} \text{ is Covariance }$ <br>
Here,  $\mu$  denotes the mean of the corresponding feature <br>
$\sigma_{xy}$ is the covariance between the features xand y, and $\sigma_{x}$    and $\sigma_{y}$  are the two features' respective standard deviations.
Variance of a feature: It simply means <i>how much below or above its mean, does the feature vary </i>
